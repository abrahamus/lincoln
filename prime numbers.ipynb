```python
def prime_list (n):

  primes = [2]

  [primes.append (x) for x in range (3, n + 1, 2) if len ([y for y in primes if x % y == 0]) == 0]

  return primes
```

```python
%%timeit
prime_list(9999)
```

```python
def prime_list1(n):
    prime_lst = [x for x in range (n + 1) if len ([y for y in range(1, x + 1) if x % y == 0]) == 2]    
    
    return prime_lst
```

```python
%%timeit
prime_list1(9999)
```

```python
def prime_list2 (n):
    prime_lst_gen = filter(lambda x: len ([y for y in range(2,x) if x % y == 0]) == 0, range(2,n+1))
    
    return list(prime_lst_gen)
```

```python
%%timeit
prime_list2(9999)
```

```python
def prime_lst (n):
    rng = n
    while True:
        lst = prime_lst(rng)
        if len(lst) == n:
            return lst
            break
        elif len(lst) < n:
            rng += 1
            continue
        else:
            rng -= 1
            continue
```
