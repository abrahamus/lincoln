```python
suits = ['hearts', 'spades', 'diamonds', 'clubs']

ranks = ['two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten', 'jack', 'queen', 'king', 'ace']

#values = dict ( zip ( ranks, range (2,15)))

values = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10, 'jack': 11, 'queen': 12, 'king': 13, 'ace': 14}
```

```python
class Card ():
    
    def __init__(self, suit, rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values [rank]
    
    def __str__(self):
        
        return f'{self.rank} of {self.suit}'
```

```python
from random import shuffle
    
class Deck ():
    
    def __init__(self):
        
        self.all = [Card (s, r) for s in suits for r in ranks]
    
    def __len__(self):
        
        return len(self.all)
    
    def __str__(self):
        
        [print(c) for c in self.all]
    
    def deal_one (self):
        
        return self.all.pop()
    
    def add_one (self, card):
        
        if type(card) == str :
            self.all.append (card)
        elif type(card) == list:
            self.all.extend (card)
    
    def shuffle (self):
        
        shuffle(self.all)
```

```python
class Player():
    
    def __init__(self, name):
        
        self.name = name
        
        self.all = []
        
    def __str__(self):
        
        return f'{self.name} has {len(self.all)} cards'
    
    def __len__(self):
        
        return len(self.all)
    
    def deal_one (self):
        
        return self.all.pop(0)
    
    def add_card (self, card):
        
        if type(card) != list:
            
            self.all.append (card)
        
        elif type(card) == list:
            
            self.all.extend (card)
```

```python
def war_game():
    
    d1 = Deck()

    d1.shuffle()

    p1 = Player('Alice')

    p2 = Player('Baker')

    for n in range (26):

        p1.add_card(d1.deal_one())
        p2.add_card(d1.deal_one())

    counter = 0

    while True:

        counter += 1
        
        if len(p1) <= 5:
            print (f'{p1.name} has <= 5 cards; {p2.name} won')
            break
            
        elif len(p2) <= 5:
            print (f'{p2.name} has <= 5 cards; {p1.name} won')
            break
        
        else:
            pass

        pool = []

        p1_card = p1.deal_one()

        p2_card = p2.deal_one()

        pool.extend ([p1_card, p2_card])

        while True:
            
            if p1_card.value > p2_card.value:
                p1.add_card(pool)
                print (f'{p1.name} has {p1_card}; {p2.name} has {p2_card}')
                print (f'{p1.name} won round {counter}')
                print (f'{p1.name} has {len(p1)} cards')
                print (f'{p2.name} has {len(p2)} cards')
                break

            elif p1_card.value < p2_card.value:
                p2.add_card(pool)
                print (f'{p1.name} has {p1_card}; {p2.name} has {p2_card}')
                print (f'{p2.name} won round {counter}')
                print (f'{p1.name} has {len(p1)} cards')
                print (f'{p2.name} has {len(p2)} cards')
                break

            elif p1_card.value == p2_card.value:
                print (f'war at round {counter}')
                
                for n in range (4):
                    p1_card = p1.deal_one()
                    p2_card = p2.deal_one()
                    pool.extend ([p1_card, p2_card])
                
                continue
            
            else:
                pass

        if counter == 1000:
            break
```
