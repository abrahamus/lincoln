```python
suits = ['hearts', 'spades', 'diamonds', 'clubs']

ranks = ['two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten', 'jack', 'queen', 'king', 'ace']

#values = dict ( zip ( ranks, range (2,15)))

values = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10, 'jack': 11, 'queen': 12, 'king': 13, 'ace': 14}
```

```python
class Card ():
    
    def __init__(self, suit, rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values [rank]
    
    def __str__(self):
        
        return f'{self.rank} of {self.suit}'
```

```python
from random import shuffle
    
class Deck ():
    
    def __init__(self):
        
        self.all = [Card (s, r) for s in suits for r in ranks]
    
    def __len__(self):
        
        return len(self.all)
    
    def __str__(self):
        
        [print(c) for c in self.all]
    
    def deal_one (self):
        
        return self.all.pop()
    
    def add_one (self, card):
        
        if type(card) == str :
            self.all.append (card)
        elif type(card) == list:
            self.all.extend (card)
    
    def shuffle (self):
        
        shuffle(self.all)
```

```python
class Player():
    
    def __init__(self, name):
        
        self.name = name
        
        self.all = []
        
    def __str__(self):
        
        return f'{self.name} has {len(self.all)} cards'
    
    def __len__(self):
        
        return len(self.all)
    
    def deal_one (self):
        
        return self.all.pop(0)
    
    def add_card (self, card):
        
        if type(card) != list:
            
            self.all.append (card)
        
        elif type(card) == list:
            
            self.all.extend (card)
```

```python
def war_game():
    
    d1 = Deck()

    d1.shuffle()

    p1 = Player('Alice')

    p2 = Player('Baker')

    for n in range (26):

        p1.add_card(d1.deal_one())
        p2.add_card(d1.deal_one())
    
    counter = 1
    
    while True:

        if len (p1) < 10 or len (p2) < 10:
            print (f'game over at round {counter}')
            print (f'{p1.name} has {len(p1)} cards')
            print (f'{p2.name} has {len(p2)} cards')
            break
            
        p1_pool = []
        p2_pool = []

        p1_pool.append(p1.deal_one())
        p2_pool.append(p2.deal_one())


        while True:
            if p1_pool[-1].value > p2_pool[-1].value:
                p1.add_card(p1_pool)
                p1.add_card(p2_pool)
                print (f'round {counter}: {p1.name} - {p1_pool[-1]} > {p2.name} - {p2_pool[-1]}')
                break

            elif p1_pool[-1].value < p2_pool[-1].value:
                p2.add_card(p2_pool)
                p2.add_card(p1_pool)
                print (f'round {counter}: {p2.name} - {p2_pool[-1]} > {p1.name} - {p1_pool[-1]}')
                break

            elif p1_pool[-1].value == p2_pool[-1].value:
                print (f'war! at round {counter}')
                for n in range (4):
                    p1_pool.append (p1.deal_one())
                    p2_pool.append (p2.deal_one())

                continue
        
        counter += 1
```
